```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN

%matplotlib inline
```

```python
# reading the image and then converting the image from BGR to RGB 
# then displaying the image
```

```python
image = cv2.imread("avengers_age_of_ultron.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# creating the empty object of MTCNN()
# this library is used for detecting faces
```

```python
mtcnn = MTCNN()
```

```python
# detecting the faces using the image
# and then taking a single face and drawing the bounding box
# after that displaying the faces with the obtained bounding boxes
```

```python
faces = mtcnn.detect_faces(image)
draw_rect = np.copy(image)

for face in faces:
    x, y, w, h = face['box']
    draw_rect = cv2.rectangle(draw_rect, (x, y), (x+w, y+h), (255, 255, 0), 2)

plt.imshow(draw_rect)
```

```python
# now using numpy library taking the box and applying the blurr effect on that part only
# then saving the blurred image
# after that displaying it
```

```python
faces = mtcnn.detect_faces(image)
blurred = np.copy(image)

for face in faces:
    x, y, w, h = face['box']
    blurred[y:y+h, x:x+h] = cv2.blur(blurred[y:y+h, x:x+h], (15, 15))

plt.imsave("Blurred_Image.png", blurred)    
plt.imshow(blurred)
```
