```python
def plotdecisionboundary(model,train,test):
    
    x1_min,x1_max = train[:,0].min()-0.1,train[:,0].max()+0.1
    x2_min,x2_max = train[:,1].min()-0.1,train[:,1].max()+0.1
    
    xx,yy = np.meshgrid(np.arange(x1_min,x1_max,0.01),
                       np.arange(x2_min,x2_max,0.01))
    
    new_input = np.c_[xx.ravel(),yy.ravel()]
    
    y_pred = model.predict_classes(new_input).reshape(xx.shape)
#     y_pred = y_pred.reshape(xx.shape)
    
    plt.contourf(xx,yy,y_pred,cmap = plt.cm.RdYlBu,alpha = 0.7)
    plt.scatter(train[:,0],train[:,1],c = test,s = 40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
def makegrid(model,data,test):
    
    x_min,x_max = data[:,0].min()-1,data[:,0].max()+1
    y_min,y_max = data[:,1].min()-1,data[:,1].max()+1
    
    xx,yy = np.meshgrid((np.arange(x_min,x_max,0.01)),
                       np.arange(y_min,y_max,0.01))

    
    new_input = np.c_[xx.ravel(),yy.ravel()]
    prediction = model.predict(new_input).reshape(xx.shape)
    
    plt.contourf(xx,yy,prediction,cmap = plt.cm.RdYlBu,alpha = 0.7)
    plt.scatter(data[:,0],data[:,1],c = test,s = 40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```
