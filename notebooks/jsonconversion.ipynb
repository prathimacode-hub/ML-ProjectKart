```python
purple.columns = ['Player','bowled', 'caught', 'stumped', 'lbw', 'caught and bowled',
       'Total Wickets', 'Team', 'Year']

purple = purple[['Year','Player','bowled', 'caught', 'stumped', 'lbw', 'caught and bowled',
       'Total Wickets', 'Team']]
purple
```

```python
purple_dict = purple.to_dict()
```

```python
purple_cap = {i:0 for i in range(2010,2017)}
for i in range(7):
    purple_cap[purple_dict['Year'][i]] = {'Player':purple_dict['Player'][i],
                                         'Bowled':purple_dict['bowled'][i],
                                         'Caught':purple_dict['caught'][i],
                                         'Stumped':purple_dict['stumped'][i],
                                         'LBW':purple_dict['lbw'][i],
                                         'Caught and Bowled':purple_dict['caught and bowled'][i],
                                         'Total Wickets':purple_dict['Total Wickets'][i],
                                         'Team' : purple_dict['Team'][i]
                                         }
    
```

```python
pprint(purple_cap)
```

```python
purple_cap = json.dumps(purple_cap,indent=2)
with open('Purplecap','w') as outfile:
    outfile.write(purple_cap)
```
