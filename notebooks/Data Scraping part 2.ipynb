```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/COVID-19_vaccination_in_India#Andhra_Pradesh')
```

```python
touse_df = dfs[7]
touse_df.head(5)
```

```python
# This method is used to convert multidimensional index values into single indexed values!

touse_df.columns = touse_df.columns.get_level_values(0)
```

```python
touse_df
```

```python
touse_df.columns
```

```python
touse_df.columns = ['State','Population','1st dose',
       '2nd dose', 'Cumulative doses administered',
       'Percentage of people given at least one dose',
       'Percentage of people fully vaccinated']
```

```python
# retriving the data from pandas dataframe

states = list(touse_df['State'])
population = list(touse_df['Population'])
dose1 = list(touse_df['1st dose'])
dose2 = list(touse_df['2nd dose'])
bothdoses = list(touse_df['Cumulative doses administered'])
atmostonedose = list(touse_df['Percentage of people given at least one dose'])
fullyvaccinated = list(touse_df['Percentage of people fully vaccinated'])


# Little preprocessing

states = [i.lower() for i in states]
population = [int(i) for i in population[:-2]]
dose1 = [int(i) for i in dose1[:-1]]
dose2 = [int(i) for i in dose2[:-1]]
atmostonedose = [float(atmostonedose[i].split('%')[0]) for i in range(len(atmostonedose[:-2]))]
fullyvaccinated = [float(fullyvaccinated[i].split('%')[0]) for i in range(len(fullyvaccinated[:-2]))]


# creating maps
populationmap = dict(list((sorted(dict(zip(states,population)).items(),key = lambda x:x[1],reverse = True))))
dose1_map = dict(list((sorted(dict(zip(states,dose1)).items(),key = lambda x:x[1],reverse = True))))
dose2_map = dict(list((sorted(dict(zip(states,dose2)).items(),key = lambda x:x[1],reverse = True))))
atmostonedose_map = dict(list((sorted(dict(zip(states,atmostonedose)).items(),key = lambda x:x[1],reverse = True))))
fulldosemap = dict(list((sorted(dict(zip(states,fullyvaccinated)).items(),key = lambda x:x[1],reverse = True))))


def getnewmap(topstates,mapper):
    maps = {}
    for key,val in mapper.items():
        if key in list(topstates.keys()):
            maps[key] = val
            
    return maps


# segregating the top doses

topstates = dict(list(populationmap.items())[:5])
topdose1 = getnewmap(topstates,dose1_map)
topdose2 = getnewmap(topstates,dose2_map)
toponedose = getnewmap(topstates,atmostonedose_map)
topfullvaccine = getnewmap(topstates,fulldosemap)
```
