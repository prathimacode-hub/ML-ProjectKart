## Finding 4's and 6's of the batsmens!{orange caps}

```python
def returnfourcount(df):
    zeros = df.loc[(df['Batsman_Scored']==0)]
    ones = df.loc[(df['Batsman_Scored']==1)]
    twos = df.loc[(df['Batsman_Scored']==2)]
    threes = df.loc[(df['Batsman_Scored']==3)]
    fours = df.loc[(df['Batsman_Scored']==4)]
    sixes = df.loc[(df['Batsman_Scored']==6)]
    name = df['Player_Name'].unique()
    return {'Player':name[0],'Zeros':zeros.shape[0],'Ones':ones.shape[0],
            'Twos':twos.shape[0],'Threes':threes.shape[0],'Fours':fours.shape[0],'Sixes':sixes.shape[0]}


# print(returnfourcount(df_2010_orange))
    
```

```python
orange_df = [df_2010_orange,df_2011_orange,df_2012_orange,df_2013_orange,df_2014_orange,df_2015_orange,df_2016_orange]
```

```python
boundary_stats = {i:0 for i in range(2010,2017)}
count = 2010
for i in range(len(orange_df)):
    boundary_stats[count] = returnfourcount(orange_df[i])
    count+=1
    
```

```python
boundary_stats
```

```python
boundary_df = pd.DataFrame(boundary_stats)
boundary_df = boundary_df.T
boundary_df['Year'] = [i for i in range(2010,2017)]
boundary_df = boundary_df[['Year','Player','Zeros','Ones','Twos','Threes','Fours','Sixes']].reset_index()
boundary_df = boundary_df.drop(['index'],axis = 1)
boundary_df
```

```python
def favbowler(df):
    temp = df.groupby(['Bowler_name'])[['Batsman_Scored']].sum()
    temp['Bowler_name'] = temp.index
    temp = temp.sort_values('Batsman_Scored',ascending = False)
    vals = temp.head()
    runs = vals['Batsman_Scored']
    names = vals['Bowler_name']
    mapper = dict(zip(names,runs))
    return mapper
```
