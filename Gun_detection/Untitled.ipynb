# To display load the trained model and detect the Gun

```python
%matplotlib inline
import torch
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Load the trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp8/weights/best.pt')

# Function to load and process image
def process_image(image_path):
    image = Image.open(image_path)
    return image

# Function for detection and drawing bounding boxes
def detect_and_show(image_path):
    image = process_image(image_path)
    results = model(image_path)

    # Convert results to Pandas DataFrame
    df = results.pandas().xyxy[0]

    fig, ax = plt.subplots(1, figsize=(12, 9))
    ax.imshow(image)

    # Draw bounding boxes for all detections since we are not using class names
    for index, row in df.iterrows():
        xmin, ymin, xmax, ymax = row['xmin'], row['ymin'], row['xmax'], row['ymax']
        rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin,
                                 linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        
        # Display confidence score near the bounding box
        ax.text(xmin, ymin - 10, f"Confidence: {row['confidence']:.2f}", color='red', fontsize=12)

    plt.show()

detect_and_show('dataset/test/images/5_jpg.rf.e062d1f2449365892f81db2a6040f70a.jpg')
```
